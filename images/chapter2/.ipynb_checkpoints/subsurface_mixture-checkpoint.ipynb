```python
from scipy.optimize import curve_fit
import scipy.stats as ss
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
errs = np.load("rec_errs.npy") + 2
nb_bins=1000
ppb = len(errs) / nb_bins
histo = plt.hist(errs-2, bins=nb_bins)
```

```python
def pretty_plot(x, errs, y, name="None"):

    f,(ax,ax2, ax3) = plt.subplots(1,3,sharey=True, facecolor='w')

    # plot the same data on both axes
    ax.plot(x, y)
    ax2.plot(x, y)
    ax3.plot(x, y)

    ax.hist(errs, bins=nb_bins)
    ax2.hist(errs, bins=nb_bins)
    ax3.hist(errs, bins=nb_bins)

    ax.set_xlim(-2,-1.96)
    ax2.set_xlim(-0.15,0.15)
    ax3.set_xlim(1.96, 2)

    # hide the spines between ax and ax2
    ax.spines['right'].set_visible(False)
    ax2.spines['left'].set_visible(False)
    ax2.spines['right'].set_visible(False)
    ax3.spines['left'].set_visible(False)
    #ax3.spines['right'].set_visible(False)
    ax.yaxis.tick_left()
    ax2.tick_params(right=False, left=False)
    ax3.tick_params(right=False, left=False)

    d = .025 

    kwargs = dict(transform=ax.transAxes, color='k', clip_on=False)
    ax.plot((1 - d, 1 + d), (-d, +d), **kwargs)  
    ax.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs) 

    kwargs.update(transform=ax2.transAxes) 
    ax2.plot((-d, +d), (-d, +d), **kwargs)       
    ax2.plot((-d, +d), (1 - d, 1 + d), **kwargs) 
    ax2.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs) 
    ax2.plot((1 - d, 1 + d), (-d, +d), **kwargs)  

    kwargs.update(transform=ax3.transAxes)
    ax3.plot((-d, +d), (-d, +d), **kwargs) 
    ax3.plot((-d, +d), (1 - d, 1 + d), **kwargs) 

    if name:
        plt.savefig(name)
    else:
        plt.show()
    
        
```

```python
def fit_function(points, sigma, lambda1, lambda2):
    normal_pdf = ss.norm.pdf(points,  loc=2, scale=sigma)
    exp1_pdf = ss.expon.pdf(points, loc=0, scale=1/lambda1)
    exp2_pdf = ss.expon.pdf(4-points, loc=0, scale=1/lambda2)

    return 0.95*ppb * normal_pdf +  0.025*ppb * exp1_pdf + 0.025*ppb * exp2_pdf
```

```python
x = np.linspace(-1, 5, len(errs))
pretty_plot(x-2, errs-2, 2*fit_function(x, .01,  100, 100))
```

```python
bins = np.linspace(-2, 2.1, nb_bins+1)
data, xbins = np.histogram(errs, bins=bins)
binscenters = np.array([0.5 * (bins[i] + bins[i+1]) for i in range(len(bins)-1)])
```

```python
popt, pcov = curve_fit(fit_function, xdata=binscenters, ydata=data, p0=[.005,  100, 100], maxfev=10000, method='trf')
popt
```

```python
pretty_plot(x-2, errs-2, 2*fit_function(x,  *popt))
```

```python
popt2, pcov2 = curve_fit(lambda points, s: ppb*ss.norm.pdf(points, loc=2, scale=s), xdata=binscenters, ydata=data, p0=[.005], maxfev=10000, method='trf')
popt2
```

```python
pretty_plot(x-2, errs-2,2*ppb*ss.norm.pdf(x, loc=2, scale=popt2[0]))
```

```python
pretty_plot(x-2, errs-2, 2*fit_function(x,  *popt), "Mixture.png")
pretty_plot(x-2, errs-2,2*ppb*ss.norm.pdf(x, loc=2, scale=popt2[0]), "Gaussian.png")
```
